```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('ecommerce_data.csv')
```

```python
df.head()
```

```python
conversion_rate = df["Comprou"].mean() * 100
print(f"\nTaxa de conversão geral: {conversion_rate:.2f}%")
```

```python
print("\nTaxa de conversão por dispositivo:")
print(df.groupby("Dispositivo")["Comprou"].mean() * 100)

print("\nTaxa de conversão por fonte de tráfego:")
print(df.groupby("Fonte_tráfego")["Comprou"].mean() * 100)

print("\nTaxa de conversão por localização:")
print(df.groupby("Localização")["Comprou"].mean() * 100)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Comprou", y="Tempo_navegação(min)")
plt.title("Relação entre tempo de navegação e conversão")
plt.xlabel("Comprou (0 = Não, 1 = Sim)")
plt.ylabel("Tempo de navegação (min)")
plt.show()
```

```python
conversion_by_age = df.groupby("Idade")["Comprou"].mean() * 100
plt.figure(figsize=(12, 6))
sns.lineplot(x=conversion_by_age.index, y=conversion_by_age.values)
plt.title("Taxa de conversão por idade")
plt.xlabel("Idade")
plt.ylabel("Taxa de conversão (%)")
plt.grid()
plt.show()
```

```python
print("\nResumo e recomendações:")
if conversion_rate < 10:
    print("- A taxa de conversão geral está baixa. Considere melhorar o design do site ou oferecer promoções.")
else:
    print("- A taxa de conversão geral está aceitável.")

print("- Investigue as fontes de tráfego e dispositivos que possuem baixa taxa de conversão para melhorias.")
```
